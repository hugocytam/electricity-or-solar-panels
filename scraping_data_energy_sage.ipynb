```python
import requests
from bs4 import BeautifulSoup
import urllib.request
import csv
URL = "https://www.energysage.com/local-data/solar-panel-cost/nm/?rc=seia"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
for points in soup.find_all("div", {"caption caption--great-price"}):
    low = str(points.text)
    print(low)
```

```python
import re
import os
temp = re.findall(r'\d+', low)
res = list(map(int, temp))
res = str(res)
res= res.replace(',', '')
res= res.replace(" ", "")
print(res)
```

```python
for points in soup.find_all("div", {"caption caption--market-average"}):
    average = str(points.text)

temp2 = re.findall(r'\d+', average)
res2 = list(map(int, temp2))
res2 = str(res2)
res2= res2.replace(',', '')
res2= res2.replace(" ", "")
print(res2)
```

```python
#az
URL = "https://www.energysage.com/local-data/solar-panel-cost/az/?rc=seia"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")

for points in soup.find_all("div", {"caption caption--great-price"}):
    low = str(points.text)
    
temp = re.findall(r'\d+', low)
temp = str(temp)
temp = temp.replace(',', '').replace(" ", "").replace("'", "")

for points in soup.find_all("div", {"caption caption--market-average"}):
    avg = str(points.text)

temp2 = re.findall(r'\d+', avg)
temp2 = str(temp2)
temp2 = temp2.replace(',', '').replace(" ", "").replace("'", "")

for points in soup.find_all("div", {"caption caption--above-market"}):
    high = str(points.text)

temp3 = re.findall(r'\d+', high)
temp3 = str(temp3)
temp3 = temp3.replace(',', '').replace(" ", "").replace("'", "")

az_values = [temp, temp2, temp3]
print(az_values)
```
